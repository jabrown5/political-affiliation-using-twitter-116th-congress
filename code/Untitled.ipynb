```python
import numpy as np
import pandas as pd
import scipy.stats as stats
import seaborn as sns
import matplotlib.pyplot as plt

pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
```

```python
# Read in data
df_115 = pd.read_csv('../data/HS115_votes.csv')
df_116 = pd.read_csv('../data/HS116_votes.csv')

df_115_mem = pd.read_csv('../data/HS115_members.csv')
df_116_mem = pd.read_csv('../data/HS116_members.csv')

df_parties = pd.read_csv('../data/HS115_parties.csv')

# https://voteview.com/data

'''
Lewis, Jeffrey B., Keith Poole, Howard Rosenthal, Adam Boche, Aaron Rudkin, and Luke Sonnet (2021). 
Voteview: Congressional Roll-Call Votes Database. https://voteview.com/
'''
```

```python
columns_keep = ['icpsr','cast_code']
columns_keep_mem = ['congress', 'chamber', 'icpsr', 'state_icpsr', 'district_code', 'state_abbrev', 'party_code', 'bioname']
columns_keep_parties = ['party_code', 'party_name']

df_115 = df_115[columns_keep]
df_116 = df_116[columns_keep]

df_115_mem = df_115_mem[columns_keep_mem]
df_116_mem = df_116_mem[columns_keep_mem]

df_parties = df_parties[columns_keep_parties]
```

```python
# HELP from: https://cmsdk.com/python/convert-a-python-dataframe-with-multiple-rows-into-one-row-using-python-pandas.html

df_115 = df_115.groupby('icpsr')['cast_code'].apply(lambda x: pd.DataFrame(x.values)).unstack().reset_index()
df_115.columns = df_115.columns.droplevel()

df_116 = df_116.groupby('icpsr')['cast_code'].apply(lambda x: pd.DataFrame(x.values)).unstack().reset_index()
df_116.columns = df_116.columns.droplevel()
```

```python
df_115.head()
```

```python
# renaming column names
# HELP: https://stackoverflow.com/questions/34049618/how-to-add-a-suffix-or-prefix-to-each-column-name
df_115 = df_115.add_prefix('115_')
df_116 = df_116.add_prefix('116_')

# rename index column
df_115.rename(columns = {'115_': 'icpsr'}, inplace = True)
df_116.rename(columns = {'116_': 'icpsr'}, inplace = True)
```
